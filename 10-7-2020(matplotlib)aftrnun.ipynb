```python
#basic linear plot
import matplotlib.pyplot as plt
x=[25,26,27,28,29,30]
y1=[28000,29000,30000,40000,50000,60000]
plt.plot(x,y1,label='y1_axis',color='y',linestyle='--',linewidth=2)
y2=[28000,32000,49000,50000,65000,80000]
plt.plot(x,y2,label='y2_axis',color='r',marker='d',linewidth=5)
plt.tight_layout()
plt.title('basic linear plot')
plt.xlabel('ages')
plt.ylabel('salary')
plt.grid(True)
plt.legend()
plt.style.use('fivethirtyeight')
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

```python
#sub plot
x=np.linspace(1,10,100)
plt.figure(1)
plt.subplot(2,2,1)
plt.title('linear')
plt.plot(x,x)

plt.subplot(2,2,2)
plt.title('cubic')
plt.plot(x,x**3)

plt.figure(2)
plt.subplot(2,2,1)
plt.title('log')
plt.plot(x,np.log(x))

plt.subplot(2,2,2)
plt.title('exponential')
plt.plot(x,x**2)
plt.show()
```

```python
# bar chart
import matplotlib.pyplot as plt
x=[25,26,27,28,29,30]
y1=[28000,29000,30000,40000,50000,60000]
plt.bar(x,y1,label='y1_axis',color='y')
plt.tight_layout()
plt.title('basic bar plot')
plt.xlabel('ages')
plt.ylabel('salary')
plt.grid(True)
plt.legend()
plt.style.use('fivethirtyeight')
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```

```python
#pie chart
import matplotlib.pyplot as plt
slices=[59267,56667,35554,78900,67890]
labels=['python','html','sql','java','css']
explode=[0,0,0,0.1,0]
plt.pie(slices,labels=labels,explode=explode)
plt.tight_layout()
plt.title('basic pie plot')
plt.grid(True)
#plt.legend()
plt.style.use('fivethirtyeight')
#print(plt.style.available)
#plt.savefig('plot.png')
plt.show()
```
