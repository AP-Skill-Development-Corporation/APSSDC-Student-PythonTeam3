## Matplotlib
- Matplotlib is Library for creating Visualization
- Matpotlib is a Python 2D Plotting

### Types of Plotting
- Line
- Bar Chart
- Histogram
- Boxplot
- Area Plot

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Line Plotting
- plt.plot()

```python
plt.plot([1,2,3,4],[1,4,6,8])# (1,1),(2,4),(3,6),(4,8)
plt.show()
```

```python
plt.plot([1,2,3,4],[1,4,6,8],color='r')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.title('Line Plotting')
```

```python
x = np.linspace(0,2,10)
x
```

```python
plt.plot(x,x,label='Line',color='g')
plt.plot(x,x**2,label='Square',color='b')
plt.plot(x,x**3,label='Cube',color='r')
plt.legend()
```

### Bar Chart
plt.bar()

```python
plt.bar([1,2,3,4],[2,6,10,12],width=0.5)
```

```python
x2 = np.arange(4)
x2
```

```python
plt.bar(x2,[90,55,40,65],width=0.3,color='g')
plt.bar(x2+0.3,[85,62,54,20],width=0.3,color='r')
```

### Histogram
- Histogram is used to show the frequency distribution
- plt.hist()

```python
a = [0,0.1,0.9,2,2.2,2.2,3,5,5,5.6]
bins = 5
plt.hist(a,bins,width = 0.4)
```

```python
a1 = [22,22.2,26,55,58,62,66,68,79,80,88,90,91,92,100]
bins = [0,25,40,70,85,105]
plt.hist(a1,bins,rwidth = 0.4)
```

### Boxplot
- Min
- IQR(Inter Quartile Range)
    - Lower Quartile(25%)
    - Median(50%)
    - Upper Quartile(75%)
- Max
- plt.boxplot()

```python
marks = np.random.randint(0,100,100)
marks
```

```python
plt.boxplot(marks)
plt.show()
```

## Area Plotting
- plt.stackplot()

```python
days = [1,2,3,4]
working = [8,8,9,10]
sleeping = [8,10,11,10]
```

```python
plt.stackplot(days,working,sleeping,colors=['g','r'])
plt.show()
```
